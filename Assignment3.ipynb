# Question 1 - Membership Functions Graphing

```python
Lam = 2
n = 2
a = 3

S = np.linspace(0,6,1000)

Ua = np.exp(-Lam * ((S-a))**2)
plt.plot(S, Ua)
plt.ylabel("Mu_a")
plt.xlabel("x")
plt.title("Membership function Mu_a ")
```

### Question 1 a)

```python
# M1 for fx = Mu_a when x <= 0.5, fx = 1-Mu_a when x > 0.5
Mu_a2 = 1 - Ua
M1 = np.minimum(Ua, Mu_a2)
plt.plot(S, Ua)
plt.plot(S, Mu_a2)

# shading the common bounded section
plt.fill_between(S, M1)
plt.title("M1 fuzziness")
plt.xlabel("X")
plt.ylabel("Mu")
plt.legend(['Mu a','1 - Mu a'])

# calculating the area under M1 using approximation
M1_result = np.trapz(M1, x=S)
print("M1 = ", M1_result)
```

### Question 1b)

```python
# M2 for alpha cut = 0.5
Mu_a_alpha = np.piecewise(S, [Ua >= 0.5, Ua < 0.5], [1,0])
M2 = np.abs(Ua - Mu_a_alpha)

plt.plot(S, Ua)
plt.plot(S, Mu_a_alpha)

# shading the common bounded section
plt.fill_between(S, Ua, Mu_a_alpha)
plt.title("M2 fuzziness")
plt.xlabel("X")
plt.ylabel("Mu")
plt.legend(['Mu a','Mu alpha'])

## calculating the area under M1 using approximation
M2_result = np.trapz(M2, x=S)
print("M2 = ", M2_result)
```

### Question 1c)

```python
# M3 for fuzzy set compliment
Mu_a_comp = 1 - Ua
M3 = np.abs(Ua - Mu_a_comp)

plt.plot(S, Ua)
plt.plot(S, Mu_a_comp)

# shading the common bounded section
plt.fill_between(S, Ua, Mu_a_comp)
plt.title("M3 fuzziness")
plt.xlabel("X")
plt.ylabel("Mu")
plt.legend(['Mu a','Mu compliment'])

## calculating the area under M1 using approximation
M3_result = np.trapz(M3, x=S)
print("M3 = ", M3_result)
```

```python
# summing values for M1, M2, and M3
MTotal = M1_result + M2_result + M3_result
print("MTotal =", MTotal)
```

# Question 2 - Characteristic of a Crisp Set Derivation

### Derivation and implications

\begin{equation*}
\chi_a = 1 - \chi_{a'}
\end{equation*}
is the compliment of the universe where 1-_a is the compliment of the set. This implies that the inverse of the set is equivalent to 1 minus the set. The implication from this result shows that the set and the compliment of the set, will encompass the entire universe, and that the compliment of the set is equivalent to the 1 minus the set.  This implies that the compliment of the set will encompass all values not found in the set.

\begin{equation*}
\chi_{a \cup b} = max[\chi_a, \chi_b]
\end{equation*}
The union of the sets a and b result in the maximum of the set being chosen, as the union operation indicates that the entire area of the two sets should be taken for all values over the span of the sets. This implies that the maximum value found in either set a or b will have the greatest effect on the result of the fuzzy data set.  

\begin{equation*}
\chi_{a \cap b} = min[\chi_a, \chi_b]
\end{equation*}  
is the intersection of the sets A and B. The resulting min function takes the two sets and compares the two, resulting in the set that is valid in both A and B. For instance, if A(x) exists, and B(x) exists, this implies that \begin{equation*}
\chi_{a \cap b}\end{equation*} exists. However, if A(x) exists, but B(x) does not, then \begin{equation*}
\chi_{a \cap b}\end{equation*} does not exist.

\begin{equation*}
\chi_{a \to b}(x,y) = min[1, {1 - \chi_a (x) + \chi_b (y)}]
\end{equation*}  
This can be shown true throug consideration of the following cases,  
if A is an element of X, but B is not an element of Y, then   
\begin{equation*}
\chi_{a \to b}(x,y) = min[1, {1 - \chi_a (x) + \chi_b (y)}] = 0
\end{equation*}  
conversely if A is an element in X, and B is an element of Y, then 
\begin{equation*}
\chi_{a \to b}(x,y) = min[1, {1 - \chi_a (x) + \chi_b (y)}]
\end{equation*}  

These results show that the operations on crisp functions will work in the same way that they will on fuzzy sets.

#### max[0, x+y-1] is non-decreasing  
for any value of x,y where x+y-1 < 0, then the max is 0 and,  
as the values of x,y increase over the period s = [0,1], then the value of max[0, x+y-1] will never decrease for increasing values of x,y.  
A specific example is when x and y are replaced with specific values, in this case a = 1, b = 1, and c = 2, d = 2, where a and c take on values for x, and b and d take on values for y. These two cases results in max[0, 1] =1 and max[0, 3] = 3 respectively. This shows that when a <= c and b <= d, then the output of the function should be non-decreasing, which it is. This will hold true for all values where a <= c and b <= d.

#### max[0, x+y-1] is commutative  
xTy = max[0, x+y-1]  
yTx = max[0, y+x-1]  
as x+y-1 = y+x-1, then xTy = yTx therefore the function is commutative.  


#### max[0, x+y-1] is associative  
For associativity (aTb)Tc = aT(bTc) two cases follow, either the inner max resolves to a + b - 1 or to 0, for the case in which the max resolves to 0 
(aTb)Tc = max[0, max[0, a + b - 1] + c - 1]

As c - 1  [-1, 0] since c  [0,1] this leads to,

For the second case when the inner max resolves to a + b - 1,

# Question 5 - Membership function for Hot, Cold, and Comfortable

### Question 5 a)

a = location where membership is greatest.
A fuzzy adjective can be described with a, as the mean understanding of noun being shifted higher or lower.  
For instance comparing hot and very hot, the very may lead individuals to place the mean value higher than the non-very counterpart (ie hot = 24, very hot = 32)  
n = detemine how crisp or fuzzy the data the function is (also known as the dilation) when determing the membership function value, for smaller values of n, the fuzziness of the data will increase, while higher values will decrease the fuziness

lambda = width of the membership function (horizontal scaling of the function)

### Question 5 b)

The Set for the data will exist in s = [10,28], where the parameters for cold, hot and comfortable are  
### Cold
a = 10, n = 3, lambda = 0.1  

### Hot
a = 28, n = 4, lambda = 0.1

### Comfortable 
a = 21, n = 1, lambda = 0.1  

The values for each greatest membership function is found from a study conducted on human health in extreme weather [1]

```python
# set up variables for each membserhip functions
Cold_a = 10
Cold_n = 3
Cold_lam = 0.1

Hot_a = 28
Hot_n = 4
Hot_lam = 0.1

Comf_a = 21
Comf_n = 1
Comf_lam = 0.1

S = np.linspace(10, 28, 1000)

# defining the membership function for the set
def Ua(a, n, Lam, x):
    return np.exp(-Lam * np.abs(x-a)**2)

Cold_Ua = Ua(Cold_a, Cold_n, Cold_lam, S)
Hot_Ua = Ua(Hot_a, Hot_n, Hot_lam, S)
Comf_Ua = Ua(Comf_a, Comf_n, Comf_lam, S)

plt.plot(S, Cold_Ua)
plt.plot(S, Hot_Ua)
plt.plot(S, Comf_Ua)
plt.ylabel("Mu_a")
plt.xlabel("x")
plt.title("Membership function Mu_a ")
plt.legend(["Cold Ua", "Hot Ua", "Comf Ua"])


```
